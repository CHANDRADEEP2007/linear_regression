```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# --- edit this to your file/columns ---
df = pd.read_csv(r"C:\Users\chand\Desktop\Exp\data_sets\Housing.csv")        # or pd.read_excel("housing.xlsx")
area_col = "area"
price_col = "price"
```

```python
# keep only the two columns and drop missing
d = df[[area_col, price_col]].dropna()
X = d[area_col].to_numpy().reshape(-1, 1)
y = d[price_col].to_numpy()
```

```python
# fit linear regression
model = LinearRegression().fit(X, y)
r2 = model.score(X, y)
```

```python
# make a smooth line for the fit
xs = np.linspace(X.min(), X.max(), 200).reshape(-1, 1)
ys = model.predict(xs)
```

```python
print(xs.shape, ys.shape, r2)
```

```python
# plot
plt.scatter(X, y, alpha=0.6, label="data")
plt.plot(xs, ys, label=f"fit: y = {model.intercept_:.2f} + {model.coef_[0]:.2f}x\nR = {r2:.3f}")
plt.xlabel(area_col)
plt.ylabel(price_col)
plt.title("Linear Regression Univariable: price ~ area")
plt.legend()
plt.tight_layout()
plt.show()
```
